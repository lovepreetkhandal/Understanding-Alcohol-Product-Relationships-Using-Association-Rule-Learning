# ASSOCIATION RULE LEARNING USING APRIORI

# importing required packages

```python
import pandas as pd
from apyori import apriori
import pickle
```

# importing the data

```python
alcohol_transactions = pd.read_csv(r"C:\Users\12176\Desktop\lab\Data science\5. Machine_learning\2. Datasets/sample_data_apriori.csv")
alcohol_transactions
```

```python
## drop the id column

alcohol_transactions.drop(["transaction_id"], axis=1, inplace=True)
```

# Modifying data for apriori algorithm


```python
### making a master list for apriori algorithm
transactions_list =[]

for index, row in alcohol_transactions.iterrows():
    transaction = list(row.dropna())
    transactions_list.append(transaction)
```

```python
transactions_list
```

# Applying the apriori algorithm

```python
apriori_rules = apriori(transactions_list,
                       min_support = 0.003,
                       min_confidence = 0.2,
                       min_lift =3,
                       min_length =2,
                       max_length =2)
```

```python
apriori_rules= list(apriori_rules)
```

```python
apriori_rules
```

# Converting output to Dataframe

```python
product1 = [list(rule[2][0][0])[0] for rule in apriori_rules]
```

```python
product1
```

```python
product2 =[list(rule[2][0][1])[0] for rule in apriori_rules]
```

```python
support = [rule[1] for rule in apriori_rules]
confidence =[rule[2][0][2] for rule in apriori_rules]
lift =[rule[2][0][3] for rule in apriori_rules]
```

```python
apriori_rules_df = pd.DataFrame({"Product_1": product1,
                                "Product_2": product2,
                                "Support": support,
                                "Confidence": confidence,
                                "Lift": lift}) 
```

```python
apriori_rules_df
```

# Sort rules by descending order

```python
apriori_rules_df.sort_values(by = "Lift", ascending =False, inplace =True)
apriori_rules_df
```
